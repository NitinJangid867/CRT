```python
# 17. Calculate the average salary using NumPy.
import numpy as np
import pandas as pd

# Load the dataset
data = pd.read_csv(r"D:\text\Book1.csv")

# Extract the Salary column
salaries = data['Salary']

# Calculate the average salary
average_salary = np.mean(salaries)

print(f"The average salary is: {average_salary}")
```

```python
# 18. Using the above Dataset , Filter out employees with salary greater than 60000 using Pandas.
import pandas as pd

# Load the dataset
data = pd.read_csv(r"D:\text\Book1.csv")
filtered_employees = data[data['Salary'] > 60000]

print(filtered_employees)
```

```python
# 19.Using the above Dataset , Plot a bar chart representing the department-wise average salary using Matplotlib.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r"D:\text\Book1.csv")

# Calculate department-wise average salary
department_avg_salary = data.groupby('Department')['Salary'].mean()

# Plot the bar chart
department_avg_salary.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Department-wise Average Salary')
plt.xlabel('Department')
plt.ylabel('Average Salary')

plt.show()
```
